```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

# Download NLTK resources
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

# Load the dataset
dataset1 = vidoni
dataset2 = liu
dataset3 = obrien

# Initialize WordNet lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Function to lemmatize words in a sentence
def lemmatize_text(text):
    tokens = word_tokenize(text.lower())
    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha() and token not in stop_words]
    return ' '.join(lemmatized_tokens)

# Lemmatize the "Comments" column
dataset1['Comments'] = dataset1['Comments'].apply(lemmatize_text)
dataset2['Comments'] = dataset2['Comments'].apply(lemmatize_text)
dataset3['Comments'] = dataset3['Comments'].apply(lemmatize_text)

# Save the modified dataset back to the same file
dataset1.to_csv("vidoni_lemma.csv", index=False)
dataset2.to_csv("liu_lemma.csv", index=False)
dataset3.to_csv("obrien_lemma.csv", index=False)


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data for LIU dataset
liu_unique_words = {
    'ALGORITHM': 3704,
    'COMPATIBILITY': 1825,
    'DEFECT': 2237,
    'DESIGN': 12249,
    'DOCUMENTATION': 978,
    'IMPLEMENTATION': 3871,
    'TEST': 1868,
    'WITHOUT_CLASSIFICATION': 11763,
    'MULTITHREAD': 75,
    'removeType': 28
}

# Data for VIDONI dataset
vidoni_unique_words = {
    'CODE': 2889,
    'DEFECT': 1990,
    'TEST': 1183,
    'ARCHITECTURE': 1271,
    'BUILD': 621,
    'REQUIREMENTS': 1180,
    'DESIGN': 810,
    'ALGORITHM': 1173,
    'DOCUMENTATION': 327,
    'VERSIONING': 93,
    'USABILITY': 291,
    'PEOPLE': 111
}

# Convert the data to DataFrame for plotting
liu_df = pd.DataFrame(liu_unique_words, index=['LIU'])
vidoni_df = pd.DataFrame(vidoni_unique_words, index=['VIDONI'])

# Set the width of the bars
bar_width = 0.35

# Get the TD types and the number of unique words for LIU and VIDONI datasets
td_types = list(liu_df.columns)
liu_counts = liu_df.iloc[0].values
vidoni_counts = vidoni_df.iloc[0].values

# Calculate the positions of the bars
liu_positions = np.arange(len(td_types))
vidoni_positions = liu_positions + bar_width

# Plot the bar graph
plt.figure(figsize=(10, 6))
liu_bars = plt.bar(liu_positions, liu_counts, bar_width, label='LIU')
vidoni_bars = plt.bar(vidoni_positions, vidoni_counts, bar_width, label='VIDONI')

# Add labels, title, and legend
plt.xlabel('TD Type')
plt.ylabel('Number of Unique Words')
plt.title('Unique Number of Words for Each TD Type')
plt.xticks(liu_positions + bar_width / 2, td_types, rotation=45)
plt.legend()

# Add space between the groups of bars
plt.subplots_adjust(bottom=0.2)

plt.tight_layout()
plt.show()
```
